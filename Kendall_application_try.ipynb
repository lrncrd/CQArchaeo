```python
from cqarchaeo import CQAnalysis, compare_quantograms
```

```python
cqa = CQAnalysis(r'data/OR_3_4.xlsx', min_value = 7, max_value = 200, min_quantum = 4, max_quantum = 24, step = 0.02, Montecarlo_sim = True, mc_parameter = 0.15, mc_iterations = 100)
```

```python
cqa.plot_quantogram(figsize=(10, 6), plot_best_quantum=True, save=False, dpi=300, plot_alpha_5=True)
```

```python
cqa_2 = CQAnalysis(r'data/OR_3_4_modified.xlsx', min_value = 7, max_value = 200, min_quantum = 4, max_quantum = 24, step = 0.02, Montecarlo_sim = True, mc_parameter = 0.15, mc_iterations = 100)
```

```python
compare_quantograms(quantogram_list = [cqa, cqa_2], figsize=(10, 6), 
                    color_list=["black", "green"], alpha_list=[0.2, 1],
                    label_list=None, plot_montecarlo_bound=[True, True])
```
